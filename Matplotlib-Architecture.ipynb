# ***Architecture of the Matplotlib library***

## **Jupyter magics**

**`%matpotlib inline`** is a **Jupyter magic** function that renders **matplotlib inside web browsers!**    
It is essentially a wrapper for the graphic device Matplotlib uses to render plots!.   
Similar to graphic devices in R e.g. png, svg, jpeg, pdf like file devices or screen devices!   
**`%matplotlib inline`** just **configures Matplotlib to be rendered inside web browsers!**  
Matplotlib has different backends that has a set of customizable configurations that decide how to render the plots in the graphic device of choice! e.g. A web browse, the OS's screen device

```python
%matplotlib inline
# inline magic is not interactive
# for every call to plot, it creates a new plot
```

`%matplotlib notebook` magic is interactive        
changes to the axes object just modifies the existing (active) artist object instead of creating new ones!

```python
import matplotlib as mpl        # importing the whole library not just pyplot
mpl.get_backend()               # check which backend is in use
```

Depending on OSes the default backend may vary.    
Frontend of Matplotlib is called the **artist layer** which is an abstraction around **drawing & layout primitives**   
**The root of visuals is a set of containers:&emsp; A figure object with one/more subplots each with a series of one/more axes**    
In general, most interactions take place with the axes object while making/modifying a plot.  
**Matplotlib figure has an `axes` object that is composed of `x`, `y` axes and separte `axis` objects for each one of them!**

**All figures are composed of numerous patch objects (pixels) that has a face colour and an edge colour.
This patch class has children called primitives! e.g shadow, wedge, arrow, ellipse, polygon, fancyarrowpatch, regularpolygon etc..**

On top of backend & artist layer is the scripting layer **`pyplot`** which actually creates the artists and augments them all together.  
Matplotlib plotting style is **procedural**. You issue a set of  instructions to the bavkend layer and it configures the graphic accordingly & renders the output!.   
In **structural** plotting libraries like the **`D3.js`**, visuals are constructed by structuring elements manually in the DOM.

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
fig.figsize = 10,10
```

```python
fig.add_axes([2,2,2,2])
```

```python
ax = plt.gca()
ax.plot(range(100))
ax.set_xlabel("Numbers")
plt.show()
```

```python
plt.figure(figsize = (10,10), dpi=200)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10,10)
fig.set_dpi(1000)
ax.set_xlabel("$X$")
ax.set_ylabel("$Y$")
ax.set_title("$Dummy\hspace{1}plot$", fontdict = {"fontsize":20, "color":"red", "fontweight":"bold"})
ax.plot(range(0, 100, 5), marker = "o", color = "yellow", linestyle='none', markersize=10, markeredgecolor="black")
plt.show()
```
