```python
# select just the 20 countries we need
asylum_in_filter = asylum_in.iloc[np.where(asylum_in.country.isin(top20.index) == True)[0], :]
asylum_in_filter = asylum_in_filter.reset_index().drop("index", axis = 1)
asylum_in_filter.head(5)
```

```python
# drop the years where more than 50% of records are NaN
np.where(asylum_in_filter.isna().mean(axis = 0) > 0.5)[0]
```

```python
asylum_in_filter.shape
```

```python
asylum_in_filter = asylum_in_filter.drop(asylum_in_filter.columns[np.where(asylum_in_filter.isna().mean(axis = 0) > 0.5)[0]], axis = 1)
asylum_in_filter.head(5)
```

```python
# it looks like all years before 2000 contain either zeroes or NaN, let's drop them as well!
asylum_in_filter = asylum_in_filter.loc[:, ["country"] + [str(i) for i in range(2000, 2021, 1)]]
asylum_in_filter
```

```python
# damn! there are strings the numeric columns
# some dumbass entered 1000s as ks!
# a function to cast these into floats!

def CastToFloat(column):
    """
    A function that takes a Pandas DataFrame column as input, checks whether that any element in that column
    contains the character "k"and if so, removes that chracter, casts the value to a float and returns it multiplied by 1000!
    If the element does not have a "k" in it, just type casts the string to a float and returns it!
    eg: "123.65k" => 123650.0
    """
    result = list()
    for value in column:
        if "k" in str(value):
            result.append(float(value.replace("k", "")) * 1000)
        else:
            result.append(float(value))
    return pd.Series(data = result, name = column.name, dtype = np.float64)
```

```python
CastToFloat(asylum_in_filter.loc[:, "2010"])
```

```python
new_asylum = asylum_in_filter[asylum_in_filter.columns[1:]].apply(CastToFloat, axis = 1)
new_asylum
```

```python
for year in new_asylum.columns:
    plt.scatter(asylum_in_filter.country, asylum_in_filter.loc[:, year])
plt.show()
```
