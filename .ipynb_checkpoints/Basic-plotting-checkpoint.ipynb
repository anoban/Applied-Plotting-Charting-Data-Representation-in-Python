```python
def randName(n):
    """
    Returns random names (arbitrary combinations of ASCII alphabets) with specified number of characters
    """
    name = ""
    sample = random.sample(ascii_letters, n)
    for i in sample:
        name += i
    return name
```

```python
doc = bs(requests.get("https://insights.stackoverflow.com/survey/2021").content, "html.parser")
```

```python
doc.find(id = "language").find_all("tr")[0].find_all("td")
```

```python
for table_row in doc.find(id = "language").find_all("tr"):
    for table_data in table_row.find_all("td"):
        print(table_data.text)
```

```python
langs = doc.find(id = "language")
```

```python
lang_stats = langs.find_all("td")
```

```python
StackOverFlow = {"Language" : [], "Percentage" : [], "Responses" : []}
for i in range(0, 76, 2):
    StackOverFlow["Language"].append(lang_stats[i].text)
    if len(lang_stats[i + 1].text.split("  ")) > 3:
        StackOverFlow["Percentage"].append(float(lang_stats[i + 1].text.split("  ")[1].replace("%", "").strip()))
        StackOverFlow["Responses"].append(int(lang_stats[i + 1].text.split("  ")[2].replace(",", "").strip()))
```

```python
Langs = pd.DataFrame(StackOverFlow)
Langs
```

```python
fig = plt.figure(figsize = (15, 7.5), dpi = 100)
axes = fig.add_subplot(111)
axes.grid(which = "both", axis = "y")
axes.set_axisbelow(True)
axes.set_ylabel("$Percentage\hspace{1}of\hspace{1}respondents$")
axes.set_xlabel("$Programming\hspace{1}language$", labelpad = 10)
axes.set_title("$Results\hspace{1}of\hspace{1}2021\hspace{1}StackOverflow\hspace{1}Survey\hspace{1}on\hspace{1}favourite\hspace{1}programming\hspace{1}languages$", fontdict = {"fontsize": 14})
axes.bar(Langs.Language, Langs.Percentage, color = "red", width = 0.75)
for tick in axes.get_xticklabels():
    tick.set_rotation(90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# change the bar color to be less bright blue
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')
# make one bar, the python bar, a contrasting color
bars[0].set_color('#1F77B4')

# soften all labels by turning grey
plt.xticks(pos, languages, alpha=0.8)
# remove the Y label since bars are directly labeled
#plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
# direct label each bar with Y axis values
for bar in bars:
    plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() - 5, str(int(bar.get_height())) + '%', 
                 ha='center', color='w', fontsize=11)
plt.show()
```
