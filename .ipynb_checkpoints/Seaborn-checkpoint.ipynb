```python
# Seaborn is essentially a wrapper around matplotlib

import pandas as pd
import seaborn 
import numpy as np
import matplotlib.pyplot as plt
importort gc
```

```python
df = pd.read_csv("D:/DataSets/Library_Usage.csv", low_memory = False)
df.head()
```

```python
fig = plt.figure(figsize = (10, 10), dpi = 75)
axes = fig.add_subplot(1,1,1)
axes.hist(df.loc[:, "Patron Type Definition"], color = "blue", align = "mid", orientation = "horizontal",
         rwidth = 0.9, bins = np.linspace(-0.5, 17.5, num = 19))
axes.tick_params(axis = "x", rotation = 90)
plt.show()
```

```python
from collections import Counter
Counter(df.loc[:, "Patron Type Definition"])
```

```python
# Renewal patterns
fig = plt.figure(figsize = (30 , 10), dpi = 50)
(axes0, axes1, axes2, axes3, axes4, axes5), (axes6, axes7, axes8, axes9, axes10, axes11), \
            (axes12, axes13, axes14, axes15, axes16, axes17) = fig.subplots(nrows = 3, ncols = 6, sharex = True, sharey = False)
grids = [axes0, axes1, axes2, axes3, axes4, axes5, axes6, axes7, axes8, axes9, axes10, axes11, \
                axes12, axes13, axes14, axes15, axes16, axes17]

visitor_types = df.loc[:, "Patron Type Definition"].unique().astype(str)
for i, axis in enumerate(grids):
    axis.hist(df.iloc[np.where(df.loc[:, "Patron Type Definition"] == visitor_types[i])[0], 3], rwidth = 0.9,
             bins = 20)
    axis.set_ylabel("Counts")
    axis.set_xlabel("Renewals")
    axis.tick_params(axis = "x", rotation = 90)
    axis.set_title(visitor_types[i])
plt.show()
```

```python
np.random.seed(2022 - 10 - 6)
```

```python
norm1 = np.random.normal(0, 10, 1000)
norm2 = np.random.normal(60, 15, 1000)
plt.hist(norm1, alpha = 0.5, bins = np.linspace(-30, 100, num = 30), color = "red", label = "norm1")
plt.hist(norm2, alpha = 0.5, bins = np.linspace(-30, 100, num = 30), color = "blue", label = "norm2")
plt.legend()
plt.show()
```

```python
plt.hist([norm1, norm2], color = ["red", "blue"],  alpha = 0.5, bins = np.linspace(-30, 100, num = 30), \
         label = ["norm1", "norm2"], rwidth = 0.9, density = True)
plt.legend()
seaborn.kdeplot(np.concatenate((norm1, norm2)))
plt.show()
```

## ***Kdeplot is a kernel density plot that gives the probability density plot of the cocatenated array!***

```python
seaborn.displot(np.concatenate((norm1, norm2)), color = "navy", kind = "kde")
seaborn.displot(np.concatenate((norm1, norm2)), color = "green", kind = "hist")
plt.show()
```

```python
seaborn.jointplot(x = df.loc[:, "Total Checkouts"][:10000],
                  y = df.loc[:, "Total Renewals"][:10000], alpha = 0.65)
plt.show()
```

```python
seaborn.jointplot(x = norm1, y = norm2, alpha = 0.5)
plt.show()
```

```python
grid = seaborn.jointplot(x = np.random.poisson(lam = 0.95, size = 10000), \
                        y = np.random.normal(loc = 1.8, scale = 0.25, size = 10000), alpha = 0.6)
grid.ax_joint.set_aspect("equal")
plt.show()
```

```python
seaborn.jointplot(x = norm1, y = norm2, alpha = 0.5, kind = "hex")
plt.show()
```

```python
seaborn.set_style("white")
```

```python
seaborn.jointplot(x = norm1, y = norm2, kind = "kde", space = 0)
plt.show()
```
