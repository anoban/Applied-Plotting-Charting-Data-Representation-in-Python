# ***Interactivity***

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
data = np.random.rand(10)
plt.plot(data)

def onClick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title("Event at pixels {}, {} and data {}, {}".format(event.x, event.y, event.xdata, event.ydata))

plt.gcf().canvas.mpl_connect("button_press_event", onClick)
```
