```python
df[(df.year == 1990) & (df.spi > -5)].plot.scatter("countyfips", "spi", c = "spi", cmap = "viridis");
```

```python
axes = df[(df.year == 1990) & (df.spi > -5)].plot.scatter("countyfips", "spi", c = "spi", cmap = "viridis")
axes.set_ylabel("$SPI$")
axes.set_xlabel("$FIPS$", labelpad = 10)
axes.set_title("$SPI\hspace{1}over\hspace{1}county\hspace{1}FIPS$")
plt.show()
```
