```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig , (axes_1, axes_2) = plt.subplots(nrows = 1, ncols = 2, sharex = True, sharey = True)
axes_1.plot(np.arange(0, 10, 0.01), np.square(np.arange(0, 10, 0.01)), marker = "o", linestyle = "none",
            color = "red", label = "$A$", markersize = 1)
axes_1.set_ylabel("$Nums$")
axes_1.grid()
axes_2.plot(np.arange(0, 10, 0.01), np.power(np.arange(0, 10, 0.01), 3), marker = "o", linestyle = "none",
            color = "green", label = "$B$",  markersize = 1.5)
plt.show()
```

```python
x = np.linspace(start = - 100,stop = 100, num = 40)
plt.rcParams["figure.figsize"] = 20, 20
fig, ((axes_1, axes_2), (axes_3, axes_4)) = plt.subplots(nrows = 2, ncols = 2, sharex = True, sharey = False)
axes_1.plot(x, color = "red")
axes_2.plot(np.sin(x), color = "blue")
axes_3.plot(x + np.random.normal(loc = 20, scale = 2.5, size = 40), color = "purple")
axes_4.plot(np.square(x), color = "crimson")
plt.show()
```

```python
fig = plt.figure(figsize = (7, 7), dpi = 100)
axes = fig.add_subplot(111)
x = np.arange(0, 100, 2.5)
y = np.square(x)
y_ = np.sqrt(np.power(x, 3))
axes.plot(x, y, color = "red")
axes.plot(x, y_, color = "blue")
axes.fill_between(x = x, y1 = y, y2 = y_, color = "yellow", alpha = 0.25)
plt.show()
```

```python
import matplotlib.gridspec as gspec
```

```python
fig = plt.figure(figsize = (10, 10), dpi = 70)
grid = gspec.GridSpec(nrows = 3, ncols = 3)   # creates a 3 x 3 plotting pane!
top = plt.subplot(grid[0, 1:]) # first row, 2nd & 3rd columns
left = plt.subplot(grid[1:, 0]) # first column, crossing 2nd & 3rd rows
bottom_right = plt.subplot(grid[1:, 1:])
```

```python
x = np.random.gamma(shape = 1.25, scale = 0.125, size = 10000)
y = np.random.normal(loc = 0, scale = 1, size = 10000)

fig = plt.figure(figsize = (10, 10), dpi = 100)
grid = gspec.GridSpec(nrows = 3, ncols = 3)   # creates a 3 x 3 plotting pane!
top = plt.subplot(grid[0, 1:]) # first row, 2nd & 3rd columns
left = plt.subplot(grid[1:, 0]) # first column, crossing 2nd & 3rd rows
bottom_right = plt.subplot(grid[1:, 1:])

bottom_right.scatter(x, y, color = "red", alpha = 0.25)
top.hist(x, bins = 1000, color = "red")
left.hist(y, bins = 1000, color = "red", orientation = "horizontal")
plt.show()
```

```python
x = np.random.gamma(shape = 1.25, scale = 0.125, size = 10000)
y = np.random.normal(loc = 0, scale = 1, size = 10000)

fig = plt.figure(figsize = (10, 10), dpi = 100)
grid = gspec.GridSpec(nrows = 3, ncols = 3)   # creates a 3 x 3 plotting pane!
top = plt.subplot(grid[0, 1:]) # first row, 2nd & 3rd columns
left = plt.subplot(grid[1:, 0]) # first column, crossing 2nd & 3rd rows
bottom_right = plt.subplot(grid[1:, 1:])

bottom_right.scatter(x, y, color = "red", alpha = 0.25)
top.hist(x, bins = 1000, color = "red")
left.invert_xaxis()
left.hist(y, bins = 1000, color = "red", orientation = "horizontal")
plt.show()
```

# ***Boxplots***

```python
rnorm = np.random.normal(loc = 1, scale = 1, size = 10000)
rgamma = np.random.gamma(shape = 2, size = 10000)
random = np.random.random(size = 10000)

fig = plt.figure(figsize = (5, 8), dpi = 100)
axes = fig.add_subplot(111) 
axes.boxplot([rnorm, rgamma, random], vert = True)
plt.show()
```

```python
import mpl_toolkits.axes_grid1.inset_locator as mpl_il
```

```python
fig = plt.figure(figsize = (5, 8), dpi = 100)
axes = fig.add_subplot(111) 
axes.boxplot([rnorm, rgamma, random], vert = True)

inset_grid = mpl_il.inset_axes(axes, width = "50%", height = "20%", loc = 1)
inset_grid.hist(rgamma)
inset_grid.yaxis.tick_right()
plt.show()
```

# ***Animations***

```python
import matplotlib.animation as anim
import time
```

**Animations are brought upon by iteratively calling a plottin function**

```python
n = 1
plt.ylabel("Frequency")
plt.xlabel("Values")
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (5, 5))

while n < 100:
    x = np.random.sample(n)   
    axes.hist(x)
    axes.set_title("Sampling {} values from the normal distribution!".format(n))
    # fig.canvas.draw()
    n += 1
    # fig.canvas.flush_events()
    plt.show()
    time.sleep(100)
```
